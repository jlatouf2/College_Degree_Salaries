```python
#drops the last column
clean_df = df.dropna()
clean_df.tail()
```

```python
#Find College Major with Highest Starting Salaries
clean_df['Starting Median Salary']#.sort_values()
```

```python
clean_df['Starting Median Salary'].idxmax()
```

```python
#shows that physician assitant is best
clean_df.sort_values(by='Starting Median Salary', ascending=False).head()
```

```python
#shows major with highest mid careers salary
clean_df.sort_values(by='Mid-Career Median Salary', ascending=False).head()
#chemical engineering
```

```python
#college major with lowest starting salary:
clean_df.sort_values(by='Starting Median Salary', ascending=True).head()
#spanish
```

```python
#which major has lowest mid career salary:
clean_df.sort_values(by='Mid-Career Median Salary', ascending=True).head()
#Education and Religion are tied
```

```python
#clean_df2 = clean_df.copy()
#this creates the Spread column:
clean_df['Spread'] = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.head()
```

```python
clean_df.head()
```

```python
#sort by spread:
#clean_df.sort_values(by='Starting Median Salary', ascending=True).head()
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
#find highest values in the 90th percentile:
clean_df.sort_values(by='Mid-Career 90th Percentile Salary', ascending=False).head()
#displays Economics as highest
```

```python
#find degrees with greatest spread in salaries:
clean_df.sort_values(by='Spread', ascending=False).head()
#so economics, finance, math
```
